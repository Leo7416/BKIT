```python
import requests
import matplotlib.pyplot as mlt
def make_url(c):
    st_url = 'http://127.0.0.1:5000/'
    res = st_url + str(c)
    return res

def get_data(c):
    url = make_url(c)
    req = requests.get(url)
    return req.json()
```

```python
c_list = [5]
for c in c_list:
    print('{} чисела Фибоначчи: {}'.format(c, get_data(c)))
```

```python
num = 12
os_y = get_data(num)
os_x = list(range(1, len(os_y)+ 1))

fig = mlt.figure(figsize = (6, 6))
mlt.xlabel('Ось абцисс')
mlt.ylabel('Ось ординат')
mlt.title('Певрые {} чисела Фибоначчи'.format(len(os_y)))
mlt.bar(os_x, os_y)
mlt.show()
```
